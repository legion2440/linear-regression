# Exercise 0: Environment and libraries

Basic environment check: Python version and the imports expected by the audit.


```python
from importlib import import_module, metadata
import platform

required_modules = [
    ("jupyter", "jupyter"),
    ("numpy", "numpy"),
    ("pandas", "pandas"),
    ("matplotlib", "matplotlib"),
    ("sklearn", "scikit-learn"),
]

python_version = platform.python_version()
major, minor, *_ = map(int, python_version.split("."))
assert (major, minor) >= (3, 9), "Python >= 3.9 is required"

print(f"Python {python_version}")
for module_name, dist_name in required_modules:
    import_module(module_name)
    print(f"{module_name}: {metadata.version(dist_name)}")
```

Short engineering note: dependency versions are pinned in `requirements.txt`, and the project is expected to run from the local `.venv`.
