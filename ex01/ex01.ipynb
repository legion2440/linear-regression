# Exercise 1: Scikit-learn estimator

The required audit output comes first. After that, the notebook adds a simple plot with the input points, the fitted line, and the prediction for `x = 4`.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.style.use("seaborn-v0_8-whitegrid")

X = np.array([[1.0], [2.1], [3.0]])
y = np.array([[1.0], [2.0], [3.0]])
x_pred = np.array([[4.0]])

model = LinearRegression().fit(X, y)
prediction = model.predict(x_pred)

print(repr(prediction))
```

```python
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
print("Score: ", model.score(X, y))
```

The line is almost perfectly aligned with the three points, and the highlighted prediction makes the result easier to inspect than a bare `score` value.


```python
fitted_x = np.linspace(X.min() - 0.2, x_pred.item() + 0.2, 200).reshape(-1, 1)
fitted_y = model.predict(fitted_x)

fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X.ravel(), y.ravel(), s=90, color="#2a9d8f", label="Observed points", zorder=3)
ax.plot(fitted_x.ravel(), fitted_y.ravel(), color="#e76f51", linewidth=2.5, label="Fitted line")
ax.scatter(x_pred.ravel(), prediction.ravel(), s=120, color="#264653", marker="X", label="Prediction for x=4", zorder=4)

ax.annotate(
    f"y = {prediction.item():.4f}",
    xy=(x_pred.item(), prediction.item()),
    xytext=(10, -18),
    textcoords="offset points",
    fontsize=10,
    color="#264653",
)
ax.set(title="LinearRegression on three points", xlabel="x", ylabel="y")
ax.legend()
plt.show()
```
