```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

plt.style.use("seaborn-v0_8-whitegrid")

X = np.arange(1, 21).reshape(10, -1)
y = np.arange(1, 11)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    shuffle=False,
)

print("X_train:\n", X_train)
print("\ny_train:\n", y_train)
print("\nX_test:\n", X_test)
print("\ny_test:\n", y_test)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X_train[:, 0], X_train[:, 1], s=90, color="#2a9d8f", label="Train")
ax.scatter(X_test[:, 0], X_test[:, 1], s=110, color="#e76f51", marker="s", label="Test")

for x1, x2, target in zip(X[:, 0], X[:, 1], y):
    ax.annotate(str(target), (x1, x2), textcoords="offset points", xytext=(6, 5), fontsize=9)

ax.set(title="Deterministic split with shuffle=False", xlabel="Feature 1", ylabel="Feature 2")
ax.legend()
plt.show()
```
